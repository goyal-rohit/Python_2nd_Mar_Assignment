#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a data visualization library in Python used to create static, animated, and interactive visualizations in Python programming language. It is built on NumPy arrays and its basic unit is a figure that contains one or more axes objects. Matplotlib provides a wide variety of customizable plots, ranging from simple bar charts and line graphs to complex 3D plots.

Matplotlib is used for data visualization in Python. Some common uses of Matplotlib include:

    Data exploration: Matplotlib can be used to explore data, create visualizations, and identify patterns and trends.
    Data communication: Matplotlib can be used to create high-quality visualizations for scientific publications, presentations, and reports.
    Machine learning: Matplotlib can be used to visualize the performance of machine learning algorithms and the results of experiments.
    Business intelligence: Matplotlib can be used to create interactive dashboards and business intelligence reports.
    Web development: Matplotlib can be used to create interactive visualizations for web applications.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

    Line plot: A line plot is used to display data points connected by straight line segments.
    Scatter plot: A scatter plot is used to display data points as individual dots.
    Bar plot: A bar plot is used to display data in a bar chart format, where the height of the bars represent the data values.
    Histogram: A histogram is used to display the distribution of a dataset by dividing the data into a set of intervals and counting the number of data points that fall into each interval.
    Pie chart: A pie chart is used to display the proportions of different categories in a dataset as slices of a pie.

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
##### Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of plot or graphical representation of data points in a two-dimensional coordinate system. It displays the values of two variables as points or dots on the graph. Each dot represents a single data point. 

Scatter plots are used to analyze the relationship between two variables and determine whether there is any correlation or pattern in the data. They are commonly used in fields such as statistics, finance, and biology to visualize and understand the relationships between variables.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,3))
plt.scatter(x,y)
plt.xlabel('This is the x Axis')
plt.ylabel('This is the y Axis')
plt.title('This is X vs Y on Scatter Plot')
```

#### Q3. Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

The subplot() function in Matplotlib is used to create multiple plots or subplots in a single figure. It is useful when we want to compare different data sets or visualize multiple aspects of a single data set in a single figure.

The subplot() function takes three arguments: the number of rows of subplots, the number of columns of subplots, and the index of the current subplot. We can then use the returned axes object to plot data on each subplot.

```python
fig, axs = plt.subplots(2, 2)


axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')


axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')


axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')


axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

plt.subplots_adjust(wspace=0.4, hspace=0.4)
plt.show()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot, also known as a box and whisker plot, is a graphical representation of a dataset's summary statistics. It displays the median, quartiles, and extremes of the data in a compact way. The box in the middle represents the interquartile range (IQR), which contains the middle 50% of the data. The whiskers extend from the box to show the range of the data, and any outliers are shown as individual points.

Box plots are useful for identifying the spread and skewness of a dataset, as well as identifying potential outliers. They are often used in statistical analysis and data exploration to summarize and visualize the distribution of a variable. Box plots are also used to compare the distributions of multiple variables or groups.

```python
data = [box1,box2]
fig, ax = plt.subplots()
ax.boxplot(data)
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Values')
ax.set_title('Box Plot Example')
```
